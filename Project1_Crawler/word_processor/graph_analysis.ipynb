```python
# Purpose: Apply Zipf's Law and Heap's Law to web crawled data
# Accepts English, Spanish, and lastly Farsi(non-latin script)

import processor as wp
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import collections
from collections import Counter
import numpy as np


english = wp.process_text("english", "../repository/English/")
spanish = wp.process_text("spanish", "../repository/Spanish/")
farsi = wp.process_text("farsi", "../repository/Farsi/")

#unique_words = set(english)
#unique_word_count = len(unique_words)
#unique_word_count
```

# English

```python
#get the unique word count 
unique_english = set(english)
unique_english_count = len(unique_english)
print('This is the english unique word count')
unique_english_count 
```

```python
df_en = pd.DataFrame(english.most_common(100),
                         columns = ['words', 'Frequency'])

i =1
temp_rank = []
temp_data_english = df_en.sort_values(by=['Frequency'], ascending = False)
temp_data_english

for w in temp_data_english['Frequency']:
    temp_rank.append(i)
    i = i + 1
    
temp_data_english['Rank'] = temp_rank
temp_data_english['R x Freq'] = temp_data_english['Rank'] * temp_data_english['Frequency']
temp_data_english

df_sorted = temp_data_english.sort_values(by=['Rank'], ascending = True)

df_sorted
#df_en.drop([0], inplace=True)
#df_en.head(50)
```

```python
#creating graphs 
w = 6
h = 4
d = 110
plt.figure(figsize=(w, h), dpi=d)
x = temp_data_english['Rank']
y = temp_data_english['Frequency']
plt.title("Zipf's Law: English")
#fig = plt.figure()
#ax = plt.axes()
#plt.grid()
plt.loglog(x, y, label='Probability')
plt.xlabel('Rank (by decreasing order)', fontsize = 10)
plt.ylabel('Probability (of occurance)', fontsize = 10)
#plt.xlim([0, 2000])
#plt.ylim([0, 1])
#plt.savefig("Zipf's Law.png")
```

```python
#Heap's law 
k = 100
b = 0.60

word = set()
r = []

for w in english:
    word.add(w)
    r.append(k*len(word)**b)
    
point = [n for n in range(0, len(r))]

plt.plot(point, r, label = "Heap's Law")
plt.title("Heap's Law")
#plt.loglog(x, y, label='nWords in Vocabulary')
plt.xlabel('Words in Collection', fontsize = 10)
plt.ylabel('nWords in Vocabulary', fontsize = 10)
plt.legend()
plt.show()
```
