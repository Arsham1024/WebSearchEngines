```python
# Purpose: Apply Zipf's Law and Heap's Law to web crawled data
# Accepts English, Spanish, and lastly Farsi(non-latin script)

import processor as wp
import pandas as pd
import matplotlib.pyplot as plt

english = wp.process_text("english", "../repository/English/")
spanish = wp.process_text("spanish", "../repository/Spanish/")
farsi = wp.process_text("farsi", "../repository/Farsi/")
```

# English

```python
df_en = pd.DataFrame(english.most_common(),
                         columns = ['words', 'count'])
df_en.drop([0], inplace=True)
df_en.tail(45)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
df_en.sort_values(by='count').plot.barh(x='words',
                                                y='count',
                                                ax=ax,
                                                color="blue")
ax.set_title("English Common Crawled Words")
plt.show()
```
